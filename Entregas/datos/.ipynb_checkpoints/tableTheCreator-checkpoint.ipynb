```python
import pandas as pd

dfp = pd.read_csv("proyectos.csv")
dfr = pd.read_csv("recursos.csv", decimal=',')
dfsocios = pd.read_csv("socios.csv", decimal=',')

dfp.head()
```

```python
dfproyectos = dfp.drop(['tipo', 'región', 'mineral', 'tipo generación', 'socios'], axis=1)
dfproyectos.head()
```

```python
dfsyp500 = pd.DataFrame()

n = dfsocios['nombre']
a = dfsocios['apellido']
na = [x+' '+y for x,y in zip(n,a)]

for i, row in dfp.iterrows():
    socios = row['socios'].split(',')
    for socio in socios:
        nombre = n[na.index(socio)]
        apellido = a[na.index(socio)]
        dfsyp500 = dfsyp500.append({'nombre': nombre, 'apellido': apellido, 'proyecto': row['nombre']}, ignore_index=True)
        
dfsyp500.head()
```

```python
dfrecursos = dfr.drop(['nombre', 'región de tramitación', 'fecha_dictamen', 'ONGs que participan'], axis=1)
dfrecursos.head()
```

```python
dfpyr400 = pd.DataFrame()
p = [x for x in dfp['nombre']]

for i, row in dfr.iterrows():
    proy = row['nombre'].replace('Recurso contra ', '')
    dfpyr400 = dfpyr400.append({'nproyecto': proy, 'nrecurso': row['número']}, ignore_index=True)
dfpyr400.head()
```

```python
dfmineras = pd.DataFrame()
dfcentrales = pd.DataFrame()

for i, row in dfp.iterrows():
    if row['tipo'] == 1:
        mins = row['mineral'].split(',')
        for min_ in mins:
            dfmineras = dfmineras.append({'nombre': row['nombre'], 'mineral': min_}, ignore_index=True)
    elif row['tipo'] == 2:
        dfcentrales = dfcentrales.append({'nombre': row['nombre'], 'tipo generación': row['tipo generación']}, ignore_index=True)

        

dfcentrales.head()
dfmineras.head()
```

```python
dftram = pd.DataFrame()

for i, row in dfr.iterrows():
    if row['status'] != 'en trámite':
        dftram = dftram.append({'número': row['número'], 'fecha_dictamen': row['fecha_dictamen']}, ignore_index=True)
dftram.head()
```

```python
dfcyr = pd.DataFrame()

comunas = []

for i, row in dfr.iterrows():
    aux = {'comuna': row['comuna'], 'region': row['region']}
    if not aux not in comunas:
        dftram = dftram.append(aux, ignore_index=True)
        comunas.append(aux)
```

```python
dfproyectos.to_csv("psql_proyectos.csv", index=False)
dfsocios.to_csv("psql_socios.csv", index=False)
dfrecursos.to_csv("psql_recursos.csv", index=False)
dfsyp500.to_csv("psql_syp.csv", index=False)
dfpyr400.to_csv("psql_ryp.csv", index=False)
dfmineras.to_csv("psql_mineras.csv", index=False)
dfcentrales.to_csv("psql_centrales.csv", index=False)
dftram.to_csv("psql_tramitados.csv", index=False)
```
