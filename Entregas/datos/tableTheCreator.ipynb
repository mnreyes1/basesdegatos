```python
import pandas as pd

dfp = pd.read_csv("proyectos.csv")
dfr = pd.read_csv("recursos.csv", decimal=',')
dfsocios = pd.read_csv("socios.csv", decimal=',')

dfsocios.rename(columns={'nombre':'snombre', 'apellido': 'apellido'}, inplace=True)
dfsocios.head()
```

```python
dfproyectos = dfp.drop(['región', 'mineral', 'tipo generación', 'socios'], axis=1)
dfproyectos.rename(columns={'nombre':'pnombre'}, inplace=True)
dfproyectos['fecha apertura'] = pd.to_datetime(dfproyectos['fecha apertura'])
for i in range(len(dfproyectos['operativo'])):
    dfproyectos['operativo'][i] = 'TRUE' if dfproyectos['operativo'][i] == 'si' else 'FALSE'
    if dfproyectos['tipo'][i] == 1:
        dfproyectos['tipo'][i] = 'minera'
    elif dfproyectos['tipo'][i] == 2:
        dfproyectos['tipo'][i] = 'central'
    elif dfproyectos['tipo'][i] == 3:
        dfproyectos['tipo'][i] = 'vertedero'
dfproyectos.head()
```

```python
dfsyp500 = pd.DataFrame()

n = dfsocios['snombre']
a = dfsocios['apellido']
na = [x+' '+y for x,y in zip(n,a)]

for i, row in dfp.iterrows():
    socios = row['socios'].split(',')
    for socio in socios:
        nombre = n[na.index(socio)]
        apellido = a[na.index(socio)]
        dfsyp500 = dfsyp500.append({'snombre': nombre, 'apellido': apellido, 'pnombre': row['nombre']}, ignore_index=True)


dfsyp500.head()
```

```python
dfrecursos = dfr.drop(['nombre', 'región de tramitación', 'fecha_dictamen', 'ONGs que participan'], axis=1)
dfr.rename(columns={'comuna de tramitación': 'comuna', 
                                 'descripción' : 'descripcion', 
                                 'área influencia (kms)': 'area',
                                 'número': 'numero'}, inplace=True)

dfrecursos['fecha apertura'] = pd.to_datetime(dfrecursos['fecha apertura'])
dfrecursos.head()
```

```python
dfpyr400 = pd.DataFrame()
p = [x for x in dfp['nombre']]

for i, row in dfr.iterrows():
    proy = row['nombre'].replace('Recurso contra ', '')
    dfpyr400 = dfpyr400.append({'pnombre': proy, 'numero': row['numero']}, ignore_index=True)
dfpyr400.head()
```

```python
dfmineras = pd.DataFrame()
dfcentrales = pd.DataFrame()

for i, row in dfp.iterrows():
    if row['tipo'] == 1:
        mins = row['mineral'].split(',')
        dfmineras = dfmineras.append({'pnombre': row['nombre'], 'mineral': mins[0]}, ignore_index=True)
    elif row['tipo'] == 2:
        dfcentrales = dfcentrales.append({'pnombre': row['nombre'], 'tipo generacion': row['tipo generación']}, ignore_index=True)

        

print(dfcentrales.head())
print(dfmineras.head())
```

```python
dftram = pd.DataFrame()

for i, row in dfr.iterrows():
    if row['status'] != 'en trámite':
        dftram = dftram.append({'numero': row['numero'], 'fecha dictamen': row['fecha_dictamen']}, ignore_index=True)


dftram['fecha dictamen'] = pd.to_datetime(dftram['fecha dictamen'])
dftram.head()
```

```python
dfcom = pd.DataFrame()

comunas = set()

for i, row in dfp.iterrows():
    if row['comuna'] not in comunas:
        dfcom = dfcom.append({'comuna': row['comuna'], 'region': row['región']}, ignore_index = True)
        comunas.add(row['comuna'])

dfcom.head()
```

```python
dfproyectos.to_csv("psql_proyectos.csv", index=False)
dfsocios.to_csv("psql_socios.csv", index=False)
dfrecursos.to_csv("psql_recursos.csv", index=False)
dfsyp500.to_csv("psql_syp.csv", index=False)
dfpyr400.to_csv("psql_ryp.csv", index=False)
dfmineras.to_csv("psql_mineras.csv", index=False)
dfcentrales.to_csv("psql_centrales.csv", index=False)
dftram.to_csv("psql_tramitados.csv", index=False)
dfcom.to_csv("psql_comunas.csv", index=False)
```
